```python
from urllib.request import urlopen
import pandas as pd
import numpy as np
```

```python
ebay="https://www.ebay.com/b/Beads-Jewelry-Making-Supplies/31723/bn_1860484"
```

```python
page = urlopen(ebay)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page)
```

```python
l=soup.find('ul', class_='b-list__items_nofooter')
l
```

```python
l.find("div").find("span",class_="s-item__price").string
```

```python
l.find("div").find("h3").string
```

```python
A=[]
for row in l.find_all("div"):
    x=row.find("h3")
    if not(x == None):
        A.append(x.string)
A
```

```python
B=[]
for row in l.find_all("div"):
    x=row.find("span",class_="s-item__price")
    if not(x == None):
        B.append(x.string)
B
```

```python
print(np.shape(A))
print(np.shape(B))
```

```python
nA=[]
for i in range(0,np.shape(A)[0],2):
    nA.append(A[i])
A=nA
A
```

```python
nB=[]
for i in range(0,np.shape(B)[0],4):
    nB.append(B[i])
B=nB
B
```

```python
print(np.shape(A))
print(np.shape(B))
```

```python
df=pd.DataFrame(A,columns=["Product"])
df["Cost"]=B
df
```

```python
df.to_csv('cost.csv', index=True)
```
